Add a column of the sum of positive_reviews + negative_reviews:

```python
data_set[['course_id','num_students']].assign(total_reviews = \
                                              data_set['positive_reviews'] + data_set['negative_reviews'])\
.head()
```

Group By: Find the courses with the large negative reviews:

```python
data_set.groupby(['course_subject'])\
.agg(nums = ('negative_reviews','sum'))\
  .sort_values('nums', ascending = False)\
  .head(7)
```

Groupby the course by "course_level", find the sum of the students, 
and sort the results:

```python
# Group By: Find the courses with the large negative reviews:
data_set.groupby(['course_level']).agg(students_mean = ('num_students', 'sum'))\
  .sort_values('students_mean', ascending = False)\
  .head()
```

Groupby the course by "course_level", find the students mean, 
and sort the results:

```python
data_set.groupby(['course_level']).agg(students_mean = ('num_students', 'mean'))\
  .sort_values('students_mean', ascending = False)\
  .head()
```

Groupby the course by "course subject", find the students mean, and sort the results:

```python
data_set.groupby(['course_subject']).agg(num_students_mean = ('num_students','mean'))\
  .sort_values('num_students_mean', ascending = False)\
  .head(10)
```
