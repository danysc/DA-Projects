```python
import pandas as pd
import numpy as np
```

Read the data from Github

```python
# Read the csv from my GitHub
df = pd.read_csv('https://raw.githubusercontent.com/danysc/PublicDataset/main/Customers-91.csv')
df.head()
```

```python
df.info()
```

```python
nRow, nCol = df.shape
```

```python
print(f'There are {nRow} rows and {nCol} columns')
```

```python
df.groupby(['Country']).agg(Coutries_nums = ('Country','count'))
```

```python
df.groupby(['Country']).agg(Coutries_nums = ('Country','count'))\
.sort_values('Coutries_nums', ascending = False)\
.head(8)
```
